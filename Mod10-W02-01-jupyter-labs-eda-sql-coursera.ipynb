## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
data = %sql select count(DISTINCT launch_site) from spacex
print('Unique launch sites: ', data[0][0])
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
data = %sql select * from spacex where (launch_site like 'CCA%') limit 5
df_cca = data.DataFrame()
print(df_cca)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
data = %sql select sum(payload_mass__kg_) from spacex where customer like '%NASA%'
print('Total payload mass carried by boosters launched by NASA:', data[0][0])
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
data = %sql select avg(payload_mass__kg_) from spacex where booster_version like '%F9 v1.1%'
print('Average payload mass carried by booster version F9 v1.1:', data[0][0])
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python

data = %sql select DATE from spacex where landing__outcome like '%ground%pad%' and mission_outcome = 'Success' order by DATE Asc limit 1
data
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
data = %sql select booster_version from spacex where (payload_mass__kg_ > 4000 and payload_mass__kg_ < 6000 and  landing__outcome like '%drone%ship%' and mission_outcome = 'Success') group by booster_version  
data
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
data = %sql select mission_outcome, count(mission_outcome) from spacex group by mission_outcome
data
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
%sql select max(payload_mass__kg_) from spacex
```

```python
data = %sql select booster_version from spacex where payload_mass__kg_ = (select max(payload_mass__kg_) from spacex) group by booster_version
data
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
%sql select booster_version, launch_site, DATE from spacex where landing__outcome like '%drone%ship%' and mission_outcome <> 'Success' and year(DATE) = 2015
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
%sql select mission_outcome, landing__outcome, count(mission_outcome) as count from spacex where DATE >= '2010-06-04' and date <='2017-03-20' group by mission_outcome, landing__outcome order by count Desc
```
